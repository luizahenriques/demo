```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo as funções
def R(Q):
    if Q > 115:
        return -480 * Q + 69000
    else:
        return 108 * Q + 1380

# Gerando valores de Q
Q_values = np.arange(0, 130, 1)

# Calculando os valores correspondentes de R(Q)
R_values = [R(Q) for Q in Q_values]

# Plotando o gráfico
plt.plot(Q_values, R_values, label='R(Q)')
plt.xlabel('Q')
plt.ylabel('R(Q)')
plt.title('Gráfico de R(Q) x Q')
plt.grid(True)
plt.legend()
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Função para calcular R(Q)
def calcular_R(Q):
    if Q > 80:
        return -480 * Q + 48000
    else:
        return 108 * Q + 960

# Gerar valores de Q
Q_valores = np.linspace(0, 130, 130)  # Valores de Q de 0 a 100

# Calcular os valores correspondentes de R(Q)
R_valores = [calcular_R(Q) for Q in Q_valores]

# Plotar o gráfico
plt.plot(Q_valores, R_valores, label='R(Q)')
plt.xlabel('Q')
plt.ylabel('R(Q)')
plt.title('Gráfico de R(Q) vs Q')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Definindo as funções de cálculo de E(R) para diferentes faixas de Q
def e_r(q):
    if q < 80:
        return 108 * q + 1300
    elif 80 <= q <= 115:
        return -3.72 * q + 10237.8
    else:
        return -480 * q + 65010

# Gerando valores de Q
q_values = np.arange(0, 130, 1)

# Calculando os valores correspondentes de E(R) para cada Q
e_r_values = [e_r(q) for q in q_values]

# Plotando o gráfico
plt.plot(q_values, e_r_values, label='E(R)', color='blue')
plt.xlabel('Q')
plt.ylabel('E(R)')
plt.title('Gráfico de E(R) em função de Q')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Letra e
# Definição da constante pw1
pw1 = 0.82

# Definição da função para calcular E(R) com base nas condições fornecidas
def e_r(q):
    if q < 80:
        return 108 * q + 420 * pw1 + 960
    elif 80 <= q <= 115:
        return -480 * q - 46620 * pw1 + 588 * pw1 * q + 48000
    else:
        return -480 * q + 21000 * pw1 + 48000

# Geração de valores de Q
q_values = np.arange(0, 150, 1)

# Cálculo dos valores correspondentes de E(R) para cada Q
e_r_values = [e_r(q) for q in q_values]

# Plotagem do gráfico
plt.plot(q_values, e_r_values, label='E(R)', color='blue')
plt.xlabel('Q')
plt.ylabel('E(R)')
plt.title('Gráfico de E(R) em função de Q')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import minimize_scalar
# Encontrar o valor de Q que maximiza E(R)
# Função negativa de E(R) para maximização
def neg_e_r(q):
    return -e_r(q)
result = minimize_scalar(neg_e_r, bounds=(0, 150), method='bounded')

# Imprimir o resultado
print("Valor de Q que maximiza E(R):", result.x)
print("Valor máximo de E(R):", -result.fun)
```
